```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from MatrixVectorizer import MatrixVectorizer
```

```python
# set a seed
np.random.seed(42)
```

```python
lr_train_df = pd.read_csv('data/lr_train.csv')
hr_train_df = pd.read_csv('data/hr_train.csv')
```

```python
print(hr_train_df.shape) # (167, 35778)
print(lr_train_df.shape) # (167, 12720)
```

```python
random_idx = np.random.randint(0, hr_train_df.shape[0])
lr_brain_vectorised = lr_train_df.iloc[random_idx].values
hr_brain_vectorised = hr_train_df.iloc[random_idx].values
```

```python
lr_brain_graph = MatrixVectorizer.anti_vectorize(lr_brain_vectorised, 160)
hr_brain_graph = MatrixVectorizer.anti_vectorize(hr_brain_vectorised, 268)
```

```python
# heatmap of the brain graph lr and hr side by side

fig, ax = plt.subplots(1, 2, figsize=(10, 5))

sns.heatmap(lr_brain_graph, ax=ax[0])
ax[0].set_title('Low Resolution Brain Graph')
ax[0].set_xlabel('Regions')
ax[0].set_ylabel('Regions')

sns.heatmap(hr_brain_graph, ax=ax[1])
ax[1].set_title('High Resolution Brain Graph')
ax[1].set_xlabel('Regions')
ax[1].set_ylabel('Regions')

plt.show()
```
